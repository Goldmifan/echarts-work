```python
import pandas as pd

data = pd.read_csv('新冠感染全球病例和死亡人数统计.csv')
data.head()
data['Cases'].sum()
```

```python
data2 = pd.read_csv('各国累计康复数据.csv')
data2.columns = ['CountryOrRegion', 'Recovered']
data2.head()
```

```python
# 以data国家为索引，将两个数据表合并
data = data.set_index('CountryOrRegion')
data2 = data2.set_index('CountryOrRegion')
data = data.join(data2)
data.fillna(-1, inplace=True)
data.to_csv('各国新冠疫情数据.csv', index=True, encoding='utf-8', sep=',')
```

```python
import pandas as pd

data = pd.read_csv('疫苗接种.csv')
data.head()
```

```python
x = data['至少接种一剂疫苗的人']
y = data['至少接种1剂的人口百分比']
y.replace('%', '', inplace=True, regex=True)
y = y.astype(float)
y /= 100
z = x / y
res = x.sum() / z.sum() * 100
res
```
